```python
def clean(s):
    s = s.strip()
    s = s.strip(",")
    s = s.strip("/")
    s = s.strip("\\")
    s = s.strip("---")
    s = s.strip()
    s = remove_non_ascii(s)
    return s
```

```python
import string
ascii = set(string.printable)   

def remove_non_ascii(s):
    st = ''
    for x in s:
        if x in ascii:
            st += x
    return st
```

### Calculating Precision / Recall / F1 metrics  

##### Threshold = 0.75 

```python
# threshold for classification
threshold = 0.75
```

```python
true_positive = 0
true_negative = 0
false_positive = 0
false_negative = 0

for i in range(len(df_processed)):
    
    s_q = df_processed['seed'][i]
    s_q = clean(s_q)
    w1 = nlp(s_q)
    
    for l in range(len(df_processed['active_titles'][i])):
        s_a = df_processed['active_titles'][i][l]
        s_a = clean(s_a)
        w2 = nlp(s_a)
        cos_sim = w1.similarity(w2)
        if cos_sim >= threshold:         ## classify as accepted match  (good) (True positive)
            true_positive += 1
            
        elif cos_sim < threshold:        ## classify as rejected match  (bad)  (False negative)
            false_negative += 1
        
    for l in range(len(df_processed['deleted_titles'][i])):
        s_d = df_processed['deleted_titles'][i][l]
        s_d = clean(s_d)
        w2 = nlp(s_d)
        cos_sim = w1.similarity(w2)
        if cos_sim >= threshold:          ## classify as accepted match   (bad)  (False positive)
            false_positive += 1
        
        elif cos_sim < threshold:          ## classify as rejected match   (good)  (True Negative)
            true_negative += 1
        
        
    if (i % 1000 == 0):
        print(i/1000)
```

```python
print("True positives: ", true_positive)
print("False positives: ", false_positive)
print("True Negatives: ", true_negative)
print("False Negatives: ", false_negative)
```

```python
Precision = true_positive / (true_positive + false_positive)
Recall =  true_positive / (true_positive + false_negative)
Accuracy = (true_positive + true_negative) / (true_positive + true_negative + false_positive + false_negative)
F1_score = (2 * Precision * Recall) / (Precision + Recall)
```

```python
print("Precision: %.3f" %Precision)
print("Recall: %.3f" %Recall)
print("Accuracy: %.3f" %Accuracy)
print("F1 Score: %.3f" %F1_score)
```

```python
docs = {}

for i in range(len(df)):
    dp = {}
    
    st = df['seed_name'].iloc[i]
    st = st.strip()
    if df['category'].iloc[i] is not None:
        st += ' '
        st += df['category'].iloc[i]
        st = st.strip()
    if df['subcategory'].iloc[i] is not None:
        st += ' '
        st += df['subcategory'].iloc[i]
        st = st.strip()
    if df['product_type'].iloc[i] is not None:
        st += ' '
        st += df['product_type'].iloc[i]
        st = st.strip()
    if df['brand'].iloc[i] is not None:
        st += ' '
        st += df['brand'].iloc[i]
        st = st.strip()
        
    st = clean(st)
    dp['seed_text'] = st       ## query concatenated text    
    
    del_text = []
    active_text = []
    
    s = len(df['detailed'][i])
    for l in range(s):
        if df['detailed'][i][l]['status'] == 'deleted':
            
            st = df['detailed'][i][l]['title']
            st = st.strip()
            if df['detailed'][i][l]['category'] is not None:
                st += ' '
                st += df['detailed'][i][l]['category']
                st = st.strip()
            if df['detailed'][i][l]['subcategory'] is not None: 
                st += ' '
                st += df['detailed'][i][l]['subcategory']
                st = st.strip()
            if df['detailed'][i][l]['prod.product_type'] is not None:
                st += ' '
                st += df['detailed'][i][l]['prod.product_type']
                st = st.strip()
            if df['detailed'][i][l]['prod.brand'] is not None:
                st += ' '
                st += df['detailed'][i][l]['prod.brand']
                st = st.strip()
                
            st = clean(st)
            del_text.append(st)   ## concatenated text of deleted prod
            
        if df['detailed'][i][l]['status'] == 'active':
            
            st = df['detailed'][i][l]['title']
            st = st.strip()
            if df['detailed'][i][l]['category'] is not None:
                st += ' '
                st += df['detailed'][i][l]['category']
                st = st.strip()
            if df['detailed'][i][l]['subcategory'] is not None:
                st += ' '
                st += df['detailed'][i][l]['subcategory']
                st = st.strip()
            if df['detailed'][i][l]['prod.product_type'] is not None:
                st += ' '
                st += df['detailed'][i][l]['prod.product_type']
                st = st.strip()
            if df['detailed'][i][l]['prod.brand'] is not None:
                st += ' '
                st += df['detailed'][i][l]['prod.brand']
                st = st.strip()
            
            st = clean(st)
            active_text.append(st)  ## concatenated text of accepted prod
    
    dp['active_text'] = active_text
    dp['deleted_text'] = del_text
    
    docs[i] = dp
```

```python
threshold_1 = 0.75
threshold_2 = 0.8
threshold_3 = 0.85
threshold_4 = 0.9
threshold_5 = 0.95

true_positive_1 = 0
true_negative_1 = 0
false_positive_1 = 0
false_negative_1 = 0

true_positive_2 = 0
true_negative_2 = 0
false_positive_2 = 0
false_negative_2 = 0

true_positive_3 = 0
true_negative_3 = 0
false_positive_3 = 0
false_negative_3 = 0

true_positive_4 = 0
true_negative_4 = 0
false_positive_4 = 0
false_negative_4 = 0

true_positive_5 = 0
true_negative_5 = 0
false_positive_5 = 0
false_negative_5 = 0

for i in range(len(df)):
    dp = {}
    
    st = df['seed_name'].iloc[i]
    st = clean(st)
    w1 = model_s.encode(st)
    
    if df['category'].iloc[i] is not None:
        st = df['category'].iloc[i]
        st = clean(st)
        t_w = model_s.encode(st)
        
        ## avg w1 & t_w
        a = np.array([w1, t_w])
        w1 = np.average(a, axis=0)
        
    if df['subcategory'].iloc[i] is not None:
        st += df['subcategory'].iloc[i]
        st = clean(st)
        t_w = model_s.encode(st)
        a = np.array([w1, t_w])
        w1 = np.average(a, axis=0)
        
        
    if df['product_type'].iloc[i] is not None:
        st += df['product_type'].iloc[i]
        st = clean(st)
        t_w = model_s.encode(st)
        a = np.array([w1, t_w])
        w1 = np.average(a, axis=0)
        
    if df['brand'].iloc[i] is not None:
        st += df['brand'].iloc[i]
        st = clean(st)
        t_w = model_s.encode(st)
        a = np.array([w1, t_w])
        w1 = np.average(a, axis=0)
         
    
    s = len(df['detailed'][i])
    for l in range(s):
        if df['detailed'][i][l]['status'] == 'deleted':
            
            st = df['detailed'][i][l]['title']
            st = clean(st)
            w2 = model_s.encode(st)
            
            if df['detailed'][i][l]['category'] is not None:
                st += df['detailed'][i][l]['category']
                st = clean(st)
                t_w = model_s.encode(st)
                a = np.array([w2, t_w])
                w2 = np.average(a, axis=0)
                
            if df['detailed'][i][l]['subcategory'] is not None: 
                st += df['detailed'][i][l]['subcategory']
                st = clean(st)
                t_w = model_s.encode(st)
                a = np.array([w2, t_w])
                w2 = np.average(a, axis=0)
                
            if df['detailed'][i][l]['prod.product_type'] is not None:
                st += df['detailed'][i][l]['prod.product_type']
                st = clean(st)
                t_w = model_s.encode(st)
                a = np.array([w2, t_w])
                w2 = np.average(a, axis=0)
                
            if df['detailed'][i][l]['prod.brand'] is not None:
                st += df['detailed'][i][l]['prod.brand']
                st = clean(st)
                t_w = model_s.encode(st)
                a = np.array([w2, t_w])
                w2 = np.average(a, axis=0)
            
            cos_sim = util.cos_sim(w1, w2)
            if cos_sim >= threshold_1:         ## classify as accepted match  (good) (True positive)
                true_positive_1 += 1
            elif cos_sim < threshold_1:        ## classify as rejected match  (bad)  (False negative)
                false_negative_1 += 1
        
            if cos_sim >= threshold_2:
                true_positive_2 += 1
            elif cos_sim < threshold_2:
                false_negative_2 += 1
            
            if cos_sim >= threshold_3:
                true_positive_3 += 1
            elif cos_sim < threshold_3:
                false_negative_3 += 1
            
            if cos_sim >= threshold_4:
                true_positive_4 += 1
            elif cos_sim < threshold_4:
                false_negative_4 += 1
            
            if cos_sim >= threshold_5:
                true_positive_5 += 1
            elif cos_sim < threshold_5:
                false_negative_5 += 1
            
            
        if df['detailed'][i][l]['status'] == 'active':
            
            st = df['detailed'][i][l]['title']
            st = clean(st)
            w2 = model_s.encode(st)
            
            if df['detailed'][i][l]['category'] is not None:
                st += df['detailed'][i][l]['category']
                st = clean(st)
                t_w = model_s.encode(st)
                a = np.array([w2, t_w])
                w2 = np.average(a, axis=0)
                
            if df['detailed'][i][l]['subcategory'] is not None: 
                st += df['detailed'][i][l]['subcategory']
                st = clean(st)
                t_w = model_s.encode(st)
                a = np.array([w2, t_w])
                w2 = np.average(a, axis=0)
                
            if df['detailed'][i][l]['prod.product_type'] is not None:
                st += df['detailed'][i][l]['prod.product_type']
                st = clean(st)
                t_w = model_s.encode(st)
                a = np.array([w2, t_w])
                w2 = np.average(a, axis=0)
                
            if df['detailed'][i][l]['prod.brand'] is not None:
                st += df['detailed'][i][l]['prod.brand']
                st = clean(st)
                t_w = model_s.encode(st)
                a = np.array([w2, t_w])
                w2 = np.average(a, axis=0)
                
            cos_sim = util.cos_sim(w1, w2)
            if cos_sim >= threshold_1:         ## classify as accepted match  (good) (True positive)
                true_positive_1 += 1
            elif cos_sim < threshold_1:        ## classify as rejected match  (bad)  (False negative)
                false_negative_1 += 1
        
            if cos_sim >= threshold_2:
                true_positive_2 += 1
            elif cos_sim < threshold_2:
                false_negative_2 += 1
            
            if cos_sim >= threshold_3:
                true_positive_3 += 1
            elif cos_sim < threshold_3:
                false_negative_3 += 1
            
            if cos_sim >= threshold_4:
                true_positive_4 += 1
            elif cos_sim < threshold_4:
                false_negative_4 += 1
            
            if cos_sim >= threshold_5:
                true_positive_5 += 1
            elif cos_sim < threshold_5:
                false_negative_5 += 1
   
    if (i % 1000 == 0):
        print(i/1000)
```
